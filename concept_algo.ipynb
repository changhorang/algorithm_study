## 배열 예제

```python
def solution(arr: list):
    arr.sort(reverse=True)
    result = list()
    for i in arr:
        if i not in result:
            result.append(i)
    
    return result

def solution2(arr: list):
    _arr = list(set(arr))
    _arr.sort(reverse=True)
    
    return _arr


arr1 = [4, 2, 2, 1, 3, 4]
print(solution2(arr1))
```

### 시간 복잡도
- N = arr length
- arr 중복 원소를 제거하는 데 걸리는 시간 복잡도 = O(N)
- 정렬하는 데 걸리는 시간 복잡도 = O(NlogN)
- 최종 시간 복잡도는 O(NlogN)

## 스택
- FILO (First In Last Out)
- push / pop / IsEmpty / IsFull
- 추상 자료형 (ADT: abstract data type)

```python
# stack 구현
# 평소에는 구현 없이 pop(), append() 메서드 이용
stack = []
max_size = 10

def isFull(stack):
    return len(stack) == max_size

def isEmpty(stack):
    return len(stack) == 0

def push(stack, item):
    if isFull(stack):
        print("FULL STACK")
    else:
        stack.append(item)
        print("SUCCESS PUSH STACK")
    
def pop(stack):
    if isEmpty(stack):
        print("EMPTY STACK")
        return None
    else:
        return stack.pop()
```

### 예제 문제
- 괄호 맞추기

```python
input = ['{', '}']
i = input.pop(0)
print(i)
```

```python
def solution(input):
    res = []
    
    for i in input: 
        if i == '(':
            res.append(i)
        else:
            res.pop()
    if len(res) == 0:
        return True
    return False

s = '(())()('
solution(s)
```

### 시간복잡도 분석
- N = s 의 length 이므로 시간복잡도 : O(N)
- append() 와 pop() 메서드는 시간복잡도 O(1)

## 10진수를 2진수로 변환하기

```python
def solution9(var):
    res = []
    
    while True:
        if var == 1:
            res.append('1')
            break
        elif var % 2 == 1:
            res.append('1')
        elif var % 2 == 0:
            res.append('0')
        var = var//2
    res.reverse()
    strt = ''.join(res)
    return strt

solution9(10)
```

### 시간 복잡도 분석
- N은 이진수 변환할 숫자
- N이 1이 될 때까지 나누므로 연산횟수는 O(logN)
- join 메서드 이용하므로 최종 시간 복잡도는 O(logN)

## 괄호 회전
- “( )”, “[ ]”, “{ }”는 모두 올바른 괄호 문자열입니다.
- 만약 A가 올바른 괄호 문자열이라면, “(A)”, “[A]”, “{A}”도 올바른 괄호 문자열입니다. 예를 들어 “[ ]”가 올바른 괄호 문자열이므로, “( [ ] )”도 올바른 괄호 문자열입니다.
- 만약 A, B가 올바른 괄호 문자열이라면, AB도 올바른 괄호 문자열입니다. 예를 들어 “{ }”와 “( [ ] )”가 올바른 괄호 문자열이므로, “{ } ( [ ] )”도 올바른 괄호 문자열입니다.
- https://school.programmers.co.kr/learn/courses/30/lessons/76502?language=python3

```python
def solution10(s):
    answer = 0
    s = list(s)
    for _ in range(len(s)):
        stack = []
        for i in range(len(s)):
            if len(stack) > 0:
                if stack[-1] == '[' and s[i] == ']': 
                    stack.pop()
                elif stack[-1] == '{' and s[i] == '}': 
                    stack.pop()
                elif stack[-1] == '(' and s[i] == ')': 
                    stack.pop()
                else:
                    stack.append(s[i])
            else:
                stack.append(s[i])
        if len(stack) == 0:
            answer += 1
            
        # 검사 후 첫 글자를 마지막으로 옮기기
        s.append(s.pop(0))
    return answer

strr = "}]()[{"
solution10(strr)
```

### 시간 복잡도 분석
- for문 2번 돌기에 O(N^2)

## 문제11 짝지어 제거하기
- 알파벳 소문자로 구성된 문자열에서 같은 알파벳이 2개 붙어 있는 짝을 찾습니다. 짝을 찾은 다음에는 그 둘을 제거한 뒤 앞뒤로 문자열을 이어붙입니다. 이 과정을 반복해서 문자열을 모두 제거한다면 짝지어 제거하기가 종료됩니다. 문자열 S가 주어졌을 때 짝지어 제거하기를 성공적으로 수행할 수 있는지 반환하는 함수를 완성하세요. 성공적으로 수행할 수 있으면 1을, 아니면 0을 반환해주면 됩니다.
- https://school.programmers.co.kr/learn/courses/30/lessons/12973

```python
def solution(s):
    stack = []
    answer = 0
    for i in s:
        # stack 이 비어있지 않고, 다음 문자열이 같은 경우
        if stack and i == stack[-1]:
            stack.pop()
        else:
            stack.append(i)
    
    # stack 에 아무것도 없는 경우 1 반환 (True)
    if not stack:
        answer = 1    

    return answer

s = 'baabaa'
solution(s)
```

### 시간 복잡도 분석
- for 문 1번 : O(N)

## 문제 12 주식 가격
- 초 단위로 기록된 주식 가격이 담긴 배열 prices가 매개변수로 주어질 때, 가격이 떨어지지 않은 기간은 몇 초인지를 반환하도록 solution( ) 함수를 완성하세요.
- prices 1 ~ 10000 / 길이는 2 ~ 100000
- https://school.programmers.co.kr/learn/courses/30/lessons/42584

```python
def solution(prices):
    n = len(prices)
    answer = [0] * n
    
    # stack 을 사용해 이전 가격과 현재가격 비교
    stack = [0] 
    for i in range(1, n):
        while stack and prices[i] < prices[stack[-1]]:
            # 가격이 떨어진 경우, 이전 가격의 기간 계산
            # print(stack)
            j = stack.pop()
            answer[j] = i - j
        stack.append(i)
    
    # stack에 남은 가격은 가격이 떨어지지 않는 경우
    while stack:
        j = stack.pop()
        answer[j] = n - 1 - j
        
    return answer

prices = [1, 2, 3, 2, 3]
# prices = [1, 6, 9, 5]
solution(prices)
```

### 분석
- index 를 활용하여 가격이 떨어지는지 확인하여 계산
- 시간 복잡도는 append / pop / for 문이므로 O(N)

```python
def solution(board, moves):
  # 각 열에 대한 스택을 생성
  lanes = [[ ] for _ in range(len(board[0]))]

  # board를 역순으로 탐색하며, 각 열의 인형을 lanes에 추가
  for i in range(len(board) - 1, -1, -1):
    for j in range(len(board[0])):
      if board[i][j]:
        lanes[j].append(board[i][j])

  # 인형을 담을 bucket을 생성
  bucket = [ ]

  # 사라진 인형의 총 개수를 저장할 변수를 초기화
  answer = 0

  # moves를 순회하며, 각 열에서 인형을 뽑아 bucket에 추가
  for m in moves:
    if lanes[m - 1]:  # 해당 열에 인형이 있는 경우
      doll = lanes[m - 1].pop( ) 

      if bucket and bucket[-1] == doll:  # 바구니에 인형이 있고, 
    # 가장 위에 있는 인형과 같은 경우
        bucket.pop( ) 
        answer += 2
    # 바구니에 인형이 없거나, 가장 위에 있는 인형과 다른 경우
      else:  
        bucket.append(doll)

  return answer
```
