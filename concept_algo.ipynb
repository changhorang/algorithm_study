# 알고리즘 공부
## 배열 예제

```python
def solution(arr: list):
    arr.sort(reverse=True)
    result = list()
    for i in arr:
        if i not in result:
            result.append(i)
    
    return result

def solution2(arr: list):
    _arr = list(set(arr))
    _arr.sort(reverse=True)
    
    return _arr


arr1 = [4, 2, 2, 1, 3, 4]
print(solution2(arr1))
```

### 시간 복잡도
- N = arr length
- arr 중복 원소를 제거하는 데 걸리는 시간 복잡도 = O(N)
- 정렬하는 데 걸리는 시간 복잡도 = O(NlogN)
- 최종 시간 복잡도는 O(NlogN)

## 스택
- FILO (First In Last Out)
- push / pop / IsEmpty / IsFull
- 추상 자료형 (ADT: abstract data type)

```python
# stack 구현
# 평소에는 구현 없이 pop(), append() 메서드 이용
stack = []
max_size = 10

def isFull(stack):
    return len(stack) == max_size

def isEmpty(stack):
    return len(stack) == 0

def push(stack, item):
    if isFull(stack):
        print("FULL STACK")
    else:
        stack.append(item)
        print("SUCCESS PUSH STACK")
    
def pop(stack):
    if isEmpty(stack):
        print("EMPTY STACK")
        return None
    else:
        return stack.pop()
```

### 예제 문제
- 괄호 맞추기

```python
input = ['{', '}']
i = input.pop(0)
print(i)
```

```python
def solution(input):
    res = []
    
    for i in input: 
        if i == '(':
            res.append(i)
        else:
            res.pop()
    if len(res) == 0:
        return True
    return False

s = '(())()('
solution(s)
```

### 시간복잡도 분석
- N = s 의 length 이므로 시간복잡도 : O(N)
- append() 와 pop() 메서드는 시간복잡도 O(1)
